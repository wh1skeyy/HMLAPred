```python
print("Unique values in purchaser_type:")
print(data['purchaser_type'].value_counts().sort_index())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure 'action_taken' is numeric for correlation
data['action_taken_encoded'] = data['action_taken'].astype('category').cat.codes

# Compute the correlation matrix
correlation_matrix = data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap (action_taken vs other variables)")
plt.show()
```
